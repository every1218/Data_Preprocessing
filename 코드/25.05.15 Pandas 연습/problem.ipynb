```python
import pandas as pd
df = pd.read_csv("C:/Users/a0102/OneDrive/바탕 화면/데이터전처리/Data_Sample.csv", header=0, encoding="utf-8")

#1
df.isnull().sum()
```

```python
#2
df = df.rename(columns={
	'Value1': 'A',
	'Value2': 'B',
	'Value3': 'C'
})
df
```

```python
#3, 4
A=df.select_dtypes(include=["number"]).columns # 수치가 포함되어 있는 변수들만
B=df.select_dtypes(include=['object', 'category']).columns # 문자와 카테고리 변수들만

df[A]=df[A].fillna(df[A].median()) 

df1 = pd.concat([df[A], df[B]], axis=1)
df1
```

```python
#5
df1['Total'] = df1['A'] + df1['B'] + df1['C']
df1.drop(['A', 'B', 'C'], axis=1, inplace=True)
df1
```

```python
#6

split_df = df['Model_Serial'].str.split('-', expand=True) # 나누고 싶은 기준 문자: '-'
# expand -> 나눠서 각 데이터프레임 열을 생성하도록 만드는 옵션

split_df.columns = ['Model', 'Serial']
split_df
```
