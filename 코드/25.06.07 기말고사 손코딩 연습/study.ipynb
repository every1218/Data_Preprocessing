```python
import pandas as pd

#1
df = pd.read_csv('C:/Users/a0102/OneDrive/바탕 화면/데이터전처리/Data_Sample.csv')
df.shape
df.columns
df.dtypes
df

#2 
df.isnull().sum()
df = df.rename(columns = {
    'Value1' : 'A',
    'Value2' : 'B',
    'Value3' : 'C'
})

#3 
df.isnull().sum()
num_vars = df.select_dtypes(include=['number'])
str_vars = df.select_dtypes(include=['object'])
num_vars =num_vars.fillna(num_vars.mean())

#4
df = pd.concat([num_vars, str_vars], axis=1)


#5 
df['total'] = df['A']+df['B']+df['C']
df=df.drop(['A','B','C'],axis=1)

#6 
split_df = df['Model_Serial'].str.split('-', expand=True)
split_df['Model_Serial'] = split_df[0]+'-'+split_df[1]
split_df.drop([0,1],axis=1)
split_df
```
