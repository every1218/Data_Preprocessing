```python
import pandas as pd 
import matplotlib.pyplot as plt 

# 한글 폰트 설정 (예: 'Malgun Gothic')
mpl.rc('font', family='Malgun Gothic')


df = pd.read_csv('C:/Users/a0102/OneDrive/바탕 화면/데이터전처리/habit_kor.csv',encoding='cp949')
df.head()
```

```python
#1 성별에 따른 시험점수 평균을 막대그래프로 시각화하시오
gender_score = df.groupby('성별')['시험점수'].mean()

gender_score.plot(kind='bar', color=['red', 'blue','green'])
plt.title('성별에 따른 시험점수 평균')
plt.xlabel('성별')
plt.ylabel('평균 시험점수')
plt.ylim(0,110)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#2 아르바이트 여부에 따른 하루 공부시간의 평균을 구하고, 이를 시각화하시오.
study = df.groupby('아르바이트여부')['하루공부시간'].mean()
study.plot(kind='bar', color = ['red', 'blue'])
plt.title('아르바이트여부에 따른 공부시간 평균')
plt.xlabel('아르바이트 여부')
plt.xlabel('하루 공부 시간')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#3 인터넷품질별 정신건강점수 평균을 선 그래프로 표현하시오.
mental = df.groupby('인터넷품질')['정신건강점수'].mean()
mental.plot(kind = 'line', color = 'skyblue')
plt.title('인터넷 품질별 정신건강점수')
plt.ylabel('정신건강점수 평균')
plt.show()
```

```python
#4 수면시간에 따른 시험점수의 산점도를 그리시오.
plt.scatter(df['수면시간'], df['시험점수'], color ='skyblue')
plt.title('수면시간에 따른 시험점수')
plt.xlabel('수면시간')
plt.ylabel('시험점수')
plt.show()
```

```python
#5 운동빈도와 정신건강점수의 관계를 산점도로 나타내고, 운동빈도가 높은 순으로 정렬한 후 시각화하시오.
plt.scatter(df['운동빈도'], df['정신건강점수'], color ='skyblue')
plt.title('운동빈도에 따른 정신건강점수')
plt.xlabel('운동빈도')
plt.ylabel('정신건강점수')
plt.show()
```
