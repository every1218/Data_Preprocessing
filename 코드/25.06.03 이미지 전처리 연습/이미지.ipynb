```python
# 1. 하나의 이미지를 다운받아 R,G,B 각각의 필터를 추출하고, 그래프로 그리시오.
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("C:/images/landscape.jpg")

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
R, G, B = cv2.split(img)

plt.figure(figsize=(8,5))
plt.subplot(1,3,1)
plt.imshow(R, cmap='Reds')
plt.title('Red')

plt.subplot(1,3,2)
plt.imshow(G, cmap='Greens')
plt.title('Green')

plt.subplot(1,3,3)
plt.imshow(B)
plt.imshow(G, cmap='Blues')
plt.title('Blue')

plt.show()
```

```python
# 2. 1번에서 다운받았던 이미지를 H, S, V 각각의 필터를 추출하고, 그래프로 그리시오.
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
hsv_img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

H,S,V = cv2.split(hsv_img)

plt.figure(figsize=(8,5))

plt.subplot(1,3,1)
plt.imshow(H, cmap = 'hsv')
plt.title('H')

plt.subplot(1,3,2)
plt.imshow(S, cmap = 'gray')
plt.title('H')

plt.subplot(1,3,3)
plt.imshow(V, cmap = 'gray')
plt.title('v')

plt.show()
```

```python
# 3. 1번에서 다운받았던 이미지를 Gray Scale로 변환하시오.
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.figure(figsize=(8,4))
plt.imshow(gray_img)
plt.title('gray scale')
plt.show()
```

```python
# 4. 1번에서 받았던 이미지를 64X64로 리사이징 한 뒤 색깔 반전 하시오.
rs_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
rs_img = 255-img
rs_img = cv2.resize(rs_img, (64, 64), interpolation=cv2.INTER_AREA)
plt.figure(figsize=(8,4))
plt.imshow(rs_img)
plt.title('resizing')
plt.show()
```

```python
# 5. 1번에서 받았던 이미지를 32X32로 리사이징 한 뒤 표준화 하시오.
rs_img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
rs_img2 = cv2.resize(rs_img2, (32,32), interpolation=cv2.INTER_AREA)
rs_img2 = rs_img2/255

plt.figure(figsize = (8,4))
plt.imshow(rs_img2)
plt.title('resizing2')
plt.show()
```

```python
# 6. random_image_test.png 파일이 어떠한 이미지인지 확인하시오.
img = cv2.imread('C:/images/random_image_test.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(8,5))
plt.imshow(img, cmap='gray')

plt.show()
```
